# 4. Newthon's method
First derivative:   

$
f' = 15(x - 3)^2\\
$
$
f'' = 30(x-3)
$




```python
import pandas as pd


func1 = lambda x: 15 * (x-3)**2
func2 = lambda x: 30 * (x-3)

def newtons_method(func1, func2, x):
    stop = 10000
    points = [x]
    f_1 = [func1(x)]
    f_2 = [func2(x)]
    while stop > 0.001:
        f_1.append(func1(x))
        f_2.append(func2(x))
        current = x - (func1(x) / func2(x))
        points.append(current)
        stop = abs(x - current)
        x = current
    return current, points, f_1, f_2

current, points, f_1, f_2 = newtons_method(func1, func2, -5)

df = pd.DataFrame({
    "func_1_vals" : f_1,
    "func_2_vals" : f_2,
    "points" : points
})

df
```

```python
func = lambda x : 5*(x-3)**3 + 4

fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(df['points'].values, func(df['points'].values))
```
